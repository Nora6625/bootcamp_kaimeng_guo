## Assumptions & Risks
- Historical distributions are assumed to continue into the future.  
- Tail events are not fully captured by variance-based risk metrics.  
- Category stability depends on external factors not modeled here.

## Sensitivity Analysis
| Scenario | Baseline | Adjusted (+20% Volatility) | Change |
|----------|----------|-----------------------------|--------|
| A        | 5.2%     | 4.4%                        | -15%   |
| B        | 6.8%     | 5.9%                        | -13%   |
| C        | 3.9%     | 3.2%                        | -18%   |

**Interpretation**: Increasing volatility by 20% reduces risk-adjusted returns substantially. Scenario B, while strong under baseline, is sensitive to volatility shifts.

## Decision Implications
- Do not rely on a single scenario; validate across alternatives.  
- Use MetricA as an early-warning KPI to monitor category divergence.  
- Next step: set up automated monitoring and periodic scenario stress testing.
